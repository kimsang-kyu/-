```python
import pandas as pd 
import geopandas as gpd
from shapely.geometry import Point, Polygon, LineString
```

```python
donggu = pd.read_csv('D:/Project/대전시 빅데이터 공모전/file/화장실정보/동구_화장실정보.csv')
yusung = pd.read_csv('D:/Project/대전시 빅데이터 공모전/file/화장실정보/유성구_화장실정보.csv')
daeduck = pd.read_csv('D:/Project/대전시 빅데이터 공모전/file/화장실정보/대덕구_화장실정보.csv')
junggu = pd.read_csv('D:/Project/대전시 빅데이터 공모전/file/화장실정보/중구_화장실정보.csv')
```

```python
#동구
donggu['lon'] = donggu['lon'].astype(float)
donggu['lat'] = donggu['lat'].astype(float)
donggu['geometry'] = donggu.apply(lambda row : Point([row['lat'], row['lon']]), axis=1)
donggu = gpd.GeoDataFrame(donggu, geometry='geometry')
```

```python
#유성구
yusung['lon'] = yusung['lon'].astype(float)
yusung['lat'] = yusung['lat'].astype(float)
yusung['geometry'] = yusung.apply(lambda row : Point([row['lat'], row['lon']]), axis=1)
yusung = gpd.GeoDataFrame(yusung, geometry='geometry')
```

```python
#대덕구
daeduck['lon'] = daeduck['lon'].astype(float)
daeduck['lat'] = daeduck['lat'].astype(float)
daeduck['geometry'] = daeduck.apply(lambda row : Point([row['lat'], row['lon']]), axis=1)
daeduck = gpd.GeoDataFrame(daeduck, geometry='geometry')
```

```python
#중구
junggu['lon'] = junggu['lon'].astype(float)
junggu['lat'] = junggu['lat'].astype(float)
junggu['geometry'] = junggu.apply(lambda row : Point([row['lat'], row['lon']]), axis=1)
junggu = gpd.GeoDataFrame(junggu, geometry='geometry')
```
